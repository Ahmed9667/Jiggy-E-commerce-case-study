Data Cleaning

```python
# Check for clumns with missing values
missing_values = df.isnull().sum().sort_values(ascending = False).reset_index()
missing_values
```

```python
missing_values = df.isnull().sum().sort_values(ascending = False).reset_index()
perc = pd.DataFrame()
perc['column'] = missing_values['index']
perc['num of missing values'] = missing_values[0]
perc
```

```python
def function(x):
    return (x/len(df)) *100
perc['percentage of missing values'] = perc['num of missing values'].apply(function)
perc
```

As we see above there are 10 columns are completely empty and 5 columns with only 10 percent of values

```python
# Handling missing values
list(perc['column'][perc['num of missing values'] > 0])
```

```python
df[list(perc['column'][perc['num of missing values'] > 0])].info()
```

```python
df.fillna({'FollowUpAction':0.0,
 'ReviewDate':0.0,
 'ReviewComment':0.0,
 'MembershipStatus':0.0,
 'CouponCode':0.0,
 'PromotionApplied':0.0,
 'CustomerSupportContact':0.0,
 'RefundAmount':0.0,
 'ReturnReason':0.0,
 'ProductReturnDate':0.0,
 'ProductDescription':0.0,
 'CustomerPhone':0.0,
 'ShipmentTrackingNumber':0.0,
 'PaymentStatus':'unknown',
 'CustomerEmail':'unknown',
 'SupplierContact':'unknown'},inplace=True)
```

```python
df.isnull().sum().sort_values(ascending=False).reset_index()
```

There are no missing values so Data is clean

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df.columns
```

```python
# Create table product
product = df[['ProductName', 'Category','ProductDescription','SupplierName', 'SupplierContact', 'PurchasePrice']].copy().drop_duplicates()

product['Product_id'] = range(1,len(product)+1)
product = product[['Product_id'] + [x for x in product.columns if x !='Product_id']]
product
```

```python
# create table customer
customer = df[['CustomerFullName', 'CustomerPhone','CustomerEmail', 'CustomerAddress'
               ,'Country', 'City', 'PostalCode']].copy().drop_duplicates()

customer['customer_id'] = range(1,len(customer)+1)
customer = customer[['customer_id']+[x for x in customer.columns if x != 'customer_id']]
customer
```

```python
# create table shipping
shipping = df[['ShippingCost', 'OrderStatus', 'ReturnStatus', 'Rating', 'Feedback',
       'ShippingDate','DeliveryDate','ShipmentTrackingNumber']].copy().drop_duplicates().reset_index(drop=True)

shipping['shipping_id'] = range(1,len(shipping)+1)
shipping = shipping[['shipping_id']+[x for x in shipping.columns if x != 'shipping_id']]
shipping
```

```python
# create table inventory
inventory = df[['Carrier', 'EstimatedDelivery',
       'ActualDelivery', 'DeliveryStatus', 'ProductReturnDate', 'ReturnReason',
       'RefundAmount', 'CustomerSupportContact', 'PromotionApplied',
       'CouponCode', 'MembershipStatus', 'ReviewComment', 'ReviewDate',
       'FollowUpAction']].copy().drop_duplicates().reset_index(drop=True)

inventory['inventory_id'] = range(1, len(inventory)+1)
inventory = inventory[['inventory_id'] + [x for x in inventory.columns if x !='inventory_id']]
inventory
```

```python
#create table sales
sales = df.merge(product , on=['ProductName', 'Category','ProductDescription','SupplierName', 'SupplierContact', 'PurchasePrice'],
                 how='left')\
       .merge(customer, on=['CustomerFullName', 'CustomerPhone','CustomerEmail', 'CustomerAddress'
               ,'Country', 'City', 'PostalCode'],how='left')\
       .merge(shipping,on=['ShippingCost', 'OrderStatus', 'ReturnStatus', 'Rating', 'Feedback',
       'ShippingDate','DeliveryDate','ShipmentTrackingNumber'],how='left')\
       .merge(inventory,on=['Carrier', 'EstimatedDelivery',
       'ActualDelivery', 'DeliveryStatus', 'ProductReturnDate', 'ReturnReason',
       'RefundAmount', 'CustomerSupportContact', 'PromotionApplied',
       'CouponCode', 'MembershipStatus', 'ReviewComment', 'ReviewDate',
       'FollowUpAction'],how='left')\
       [['Product_id','customer_id','shipping_id','inventory_id','Quantity', 'UnitPrice',
       'TotalPrice','PaymentType','Profit',
       'Discount', 'Tax', 'TaxAmount', 'FinalPrice', 'PaymentStatus','Date']]

sales
```

```python
#save tables
product.to_csv(r'clean tables\product.csv',index=False)
customer.to_csv(r'clean tables\customer.csv',index=False)
shipping.to_csv(r'clean tables\shipping.csv',index=False)
inventory.to_csv(r'clean tables\inventory.csv',index=False)
sales.to_csv(r'clean tables\sales.csv',index=False)
```
